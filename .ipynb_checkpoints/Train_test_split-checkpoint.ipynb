```python
from scipy.io import loadmat
from collections import defaultdict
import shutil
import os
```

```python
mydict = defaultdict(list)
mat = loadmat('./data/devkit/cars_annos.mat')
files = mat['annotations'][0]
```

```python
pair = []
for line in files:
    file_name = line[0][0]
    file_class = line[-2][0][0]
    pair.append((file_name,file_class))
for fname,fclass in pair:
    mydict[fclass].append(fname)
```

```python
mydict
```

```python
source_path = './data/'
train_output = './data/train_cos/'
test_output = './data/test_cos/'
if not os.path.exists(train_output):
    os.mkdir(train_output)
if not os.path.exists(test_output):
    os.mkdir(test_output)
for key in mydict.keys():
    total = len(mydict[key])
    train = int(total*0.8)
    test = total - train
    train_output_path = os.path.join(train_output,str(key))
    if not os.path.exists(train_output_path):
        os.mkdir(train_output_path)
        
    test_output_path = os.path.join(test_output,str(key))
    if not os.path.exists(test_output_path):
        os.mkdir(test_output_path)
                 
    for i in range(total):
        file_name = mydict[key][i]
     #   print(os.path.join(source_path,file_name))
        if i<=train:
            shutil.copy(os.path.join(source_path,file_name),train_output_path)
        else:
            shutil.copy(os.path.join(source_path,file_name),test_output_path)
```
